## Pattern distinguishability maximalization
We have 44 phonemes that require a pattern. These patterns need to be distinguishable from eachother, to the highest degree possible.  
Below is a model which can compare two patterns, pairwise only. At the bottom an overview of the methods of pattern collection is given.

## Pairwise Model
https://github.com/MKLab-ITI/visil provides a pretrained model for video distinguishability (referred to as ViSiL).
We shall use this model as a baseline for pairwise pattern distinguishability.

## Complete Patterns Model
The above model gives us the possibility to calculate the total score of a pair, but does not provide a method for quickly calculating the score of a collection of patterns.
There is of course the brute force method, generating a new set of patterns and getting the score for them, improving over time, but this is not practical due to runtime.
Then there are two known methods of decreasing runtime for calculating the score of a collection of patterns.
### Simulated Annealing (SA)
Simulated Annealing is a method where the score of a collection of patterns is improved by starting with a random collection of patterns, and then iteratively changing the collection of patterns by (sub)randomly changing one of the patterns.

### Genetic Algorithm (GA)
Genetic Algorithm is a method where the score of a collection of patterns is improved by combining two pattern collections by swapping half the patterns in each with the other.
Afterwards, continue with both collections using Simulated Annealing.

## Step by step plan
This plan works on a thread-by-thread basis, and can be multithreaded to speed up the process. Files with scores are ouput at the end without interaction between threads hence it is threadsafe.
1. Generate two sets of patterns.
2. Take the first of the two sets.
3. Calculate and store the score of each pair of patterns using ViSiL. The score should be stored in a hash (or similar) to be used later.
4. Calculate the score of the set. This is done by taking the median score of the pairs in the set.
5. Swap out a pair using SA.
6. Recalculate the score of the set and decide what action to undertake next using SA. See the algorithm code below.
7. First set? Go to 7.1. Second set? Go to 7.2.
   1. Take the second set and go to step 3.
   2. If the max iterations are not reached, apply GA to the two sets and continue with step 2. Otherwise, apply SA a last time before returning the best score.

### Import

```python
import numpy as np
import random
from collections.abc import Callable, Mapping, Collection
from typing import Tuple
```

### Pattern Class
Will hold some basic information about a pattern.

```python
class Pattern:
    def __init__(self, file: str):
        self.file = file
    
    def __str__(self):
        return self.file
    def __repr__(self):
        return self.__str__()
```

### Pair-wise score calculation
Caches existing scores in hashmap.
The `__pair_score_function__` is used t calculate the score of a pair of patterns. 

```python
# Modify this method to define your own pattern score calculation
def get_pair_score_function(x: Pattern, y: Pattern):
    query_video = load_video('examples/' + str(x))
    target_video = load_video('examples/' + str(y))
    
    query_features = model.extract_features(query_video, batch_sz=32)
    target_features = model.extract_features(target_video, batch_sz=32)
    return model.calculate_video_similarity(query_features, target_features)
```

```python
__score_cache__: Mapping = {} # Do not use this variable elsewhere!
def get_pattern_pair_score(x: Pattern, y: Pattern) -> np.float32:
    if (x, y) in __score_cache__:
        return __score_cache__[(x, y)]
    else:
        score = get_pair_score_function(x, y)
        __score_cache__[(x, y)] = score
        return score
```

### Complete pattern set score calculation.
Calculates the full set score (using the cached scores). Requires a Callable function to calculate the pattern score (which we suggest to be np.median).

```python
def get_set_score(element_list: Collection,score_picking: Callable) -> np.float32:
    scores = []
    for i, x in enumerate(element_list):
        for j, y in enumerate(element_list):
            if i == j:
                continue
            scores.append(get_pattern_pair_score(x, y))
            
    return 1 - abs(score_picking(scores))
```

### Dynamic Pattern Generation
Uses a pattern generator to generate new random patterns.

```python
def get_dynamic_pattern() -> Pattern:
    return Pattern('p_' + str(np.random.randint(0, 1000)) + '.gif')
```

### Simulated Annealing
Uses the dynamic pattern generator to generate new patterns.

```python
__temperature_lower_bound__ = 0.05 # Do not use elsewhere. Only modify here.
def SA(element_set: Collection, score_picking: Callable, temperature: float = 1.0,
       cooling_rate: float = 0.99, iterations: int = 100) -> Tuple[Pattern, np.float32]:
    
    # Print SA details
    print("Starting {} iterations of Simulated Annealing on set of size {}, with initial temp {}, and a cooling rate of {}"
         .format(iterations, len(element_set), temperature, cooling_rate))

    # Get the current score of the set
    current_score = get_set_score(element_set, score_picking)

    # Run through iterations
    for i in range(iterations):

        # Create a copy of the set
        new_element_set = element_set.copy()

        # Replace a pattern in the set with a new one
        new_element_set.remove(random.choice(new_element_set))
        new_element_set.append(get_dynamic_pattern())
        
        # Get the new score of the set
        new_score = get_set_score(new_element_set, score_picking)

        # If the new score is better, keep the new set
        if new_score > current_score:
            element_set = new_element_set
            current_score = new_score

        # Else accept the new set with a probability
        else:
            if np.random.random() < np.exp((new_score - current_score) / temperature):
                element_set = new_element_set
                current_score = new_score

        # Cool the temperature
        temperature *= cooling_rate
            
        # Print an update every 100 cycles
        if i % 100 == 0:
            print("(I {} / {}) \t Temp @ {} \t Score {} with set {}".format(i, iterations, temperature, new_score, element_set))
        
        # Stop if the temperature is lower than __temperature_lower_bound__
        if temperature <= __temperature_lower_bound__:
            print("(I {} / {}) \t Temp @ {} \t Temp lower than lower bound of {}. Score {} with set {}".format(i, iterations, temperature, __temperature_lower_bound__))
            break
    
    # Print summary
    print("(I {} / {}) \t Temp @ {} \t Finished iterating. Score {} with set {}")
    
            
    # Return the new pattern set and its score
    return element_set, current_score
```
